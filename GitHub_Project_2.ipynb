### Data Cleaning

```python
df = pd.read_csv('online_sales.csv')
df.info()
```

```python
pd.set_option('display.max_columns', None)
df.head()
```

```python
df.dropna()
```

```python
df['Transaction_Date'] = df['Transaction_Date'].astype('datetime64[ns]')
df.drop(['Date'], axis='columns', inplace=True)
```

### Data Analysis

```python
df['Sale'] = df['Quantity']*df['Avg_Price']
```

```python
df.to_csv('df_cleaned.csv', sep=',', index=False, encoding='utf-8')
```

```python
df_unique = df.drop_duplicates(subset=['CustomerID'])
```

```python
df_unique['Online_Perc'] = (df_unique['Online_Spend']/(df_unique['Offline_Spend']+df_unique['Online_Spend']))*100
df_unique['Online_Perc'] = df_unique['Online_Perc'].round(2)
```

```python
import numpy as np

conditions = [
    (df_unique['Online_Perc'] <= 25),
    (df_unique['Online_Perc'] > 25) & (df_unique['Online_Perc'] <= 50),
    (df_unique['Online_Perc'] > 50) & (df_unique['Online_Perc'] <= 75),
    (df_unique['Online_Perc'] > 75)
    ]

values = ['1_Q', '2_Q', '3_Q', '4_Q']

df_unique['Quartile'] = np.select(conditions, values)
df_unique.head()
```

```python
df_unique.dropna(inplace=True)
```
